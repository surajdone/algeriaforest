```python
def corelation(dataset,threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in  range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold :
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
                   
```

```python
corr_feature=corelation(x_train,0.85)
```

```python
corr_feature
```

```python
x_train.drop(corr_feature,axis=1,inplace=True)
x_test.drop(corr_feature,axis=1,inplace=True)
```

# feature scalling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaller=StandardScaler()
```

```python
x_train_scalled=scaller.fit_transform(x_train)
x_test_scalled=scaller.transform(x_test)
```

```python
x_train_scalled
```

```python
plt.subplots(figsize=(20,5))
plt.subplot(1,2,1)
sns.boxplot(data=x_train)
plt.title("Without scalled data")
plt.subplot(1,2,2)
sns.boxplot(data=x_train_scalled)
plt.title("scalled data")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
regrresor=LinearRegression()
regrresor.fit(x_train_scalled,y_train)
y_pred=regrresor.predict(x_test_scalled)
mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("Mean absolute error", mae)
print("R2 Score", score)
```

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
lasso=Lasso()
```

```python
lasso.fit(x_train_scalled,y_train)
y_pread=lasso.predict(x_test_scalled)
mae=mean_absolute_error(y_test,y_pread)
score=r2_score(y_test,y_pread)
print("the lasso mae is",mae)
print("score of r_square is",score)
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge()
ridge.fit(x_train_scalled,y_train)
y_pread=ridge.predict(x_test_scalled)
mae=mean_absolute_error(y_test,y_pread)
score=r2_score(y_test,y_pread)
print("thr ridge mae is",mae)
print("score of r_square is",score)
```
